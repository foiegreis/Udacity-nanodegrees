# Install libraries

we'll need some installed libraries to run our project. We want them to stay installed after runtime disconnection and to be placed in a comfortable path on our drive, in order to reuse them for the next projects!

##Mount gdrive and create a path symlink

this will allow you to keep the libraries installed after the runtime has shut down

**the folder called '/lib' MUST be placed outside the github folder you cloned** 
I used the path 'content/gdrive/MyDrive/GoogleColab/lib'




see:
https://colab.research.google.com/drive/1KpMDi9CjImudrzXsyTDAuRjtbahzIVjq

https://stackoverflow.com/questions/66883214/cant-find-module-in-pip-after-install-it-in-google-drive-using-google-colab

https://ayoolafelix.hashnode.dev/how-to-permanently-install-a-module-on-google-colab-ckixqrvs40su044s187y274tc

```python
import os, sys
from google.colab import drive
drive.mount('/content/gdrive')

libs_path = '/content/lib'
os.symlink('/content/gdrive/MyDrive/GoogleColab/lib', libs_path)
sys.path.insert(0, libs_path)  
```

we'll install all the libraries as

!pip install --target=$libs_path 'library'
